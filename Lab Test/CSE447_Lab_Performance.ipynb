#Task 2
Find the decrypted plaintext from the following:

n = 403212583809059293300199955599

c = 360112796689162206303023399384

e = 65537

Hint: Sympy is a great tool for complex mathematical operations

```python
# Enter your code here
from sympy import factorint

n = 403212583809059293300199955599
c = 360112796689162206303023399384
e = 65537

p, q = factorint(n)
print(p,q, (p*q))

phi_n = (p - 1) * (q - 1)

d = pow(e, -1, phi_n)

M = pow(c, d, n)
print(M)
```

#Task 3

Image: https://drive.google.com/file/d/1ea2M_L4RsJLUwGnj8LBaErxiek3CbAox/view?usp=sharing

Message: **birds_of_a_same_feather_flock_together**

Hide the message inside the green layer of the image and extract the message again to verify

```python
from PIL import Image
```

```python
# Your code here
def hide_message(image_path, message):
    img = Image.open(image_path)

    binary_message = ''.join(format(ord(i), '08b') for i in message)

    width, height = img.size

    max_message_length = (width * height) // 8

    if len(binary_message) > max_message_length:
        raise ValueError('Message too long to be hidden in the given image')

    new_img = Image.new(img.mode, img.size)

    message_index = 0

    # Loop through each pixel in the image - use nested loop where x represents width & y represents height
    for x in range(width):
      for y in range(height):

            # Get the RGB values of the current pixel, using img.getpixel()
            pixel= img.getpixel((x,y))
            red, green, blue= pixel

            # Check if the message has been fully hidden
            if message_index == len(binary_message):
                new_img.putpixel((x, y), pixel)
                continue

            # Modify the least significant bit of the green channel with the message bits
            green_binary = format(green, '08b')
            green_binary = green_binary[:-1] + binary_message[message_index]
            message_index += 1

            # Convert back to integer
            green = int(green_binary, 2)

            # Create a new pixel with modified green channel and original red & blue channels
            new_pixel = (red, green, blue)
            new_img.putpixel((x, y), new_pixel)

    new_img.save('hidden_Bird.png')
```

```python
hide_message("/content/bird.jpg", "birds of a same feather flock together")
```

```python
def extract_message(image_path):
    # Open the image
    img = Image.open(image_path)

    # Get the size of the image
    width, height = img.size

    binary_message = ''

    # Loop through each pixel in the image
    for x in range(width):
        for y in range(height):
            pixel = img.getpixel((x, y))
            red, green, blue = pixel

            # Only consider the least significant bit of the green channel
            green_binary = format(green, '08b')
            binary_message += green_binary[-1]

    # Convert the binary message to text
    message = ''
    for i in range(0, len(binary_message), 8):
        # Ensure we only decode full bytes
        byte = binary_message[i:i+8]
        if len(byte) == 8:  # Check if we have a complete byte
            message += chr(int(byte, 2))

    return message
```
