```
from PIL import Image

def hide_message(image_path, message):
    # Open the grayscale image
    img = Image.open(image_path).convert('L')

    # Convert the message to binary format
    binary_message = ''.join(format(ord(i), '08b') for i in message)

    # Get the image size
    width, height = img.size

    # Calculate the maximum message length that can fit in the image
    max_message_length = (width * height) // 8

    # Raise an error if the message is too long
    if len(binary_message) > max_message_length:
        raise ValueError('Message too long to be hidden in the given image')

    # Create a new image for the result
    new_img = Image.new(img.mode, img.size)

    message_index = 0

    # Loop through each pixel in the image
    for x in range(width):
        for y in range(height):

            # Get the grayscale value of the current pixel
            pixel = img.getpixel((x, y))

            # Check if the message has been fully hidden
            if message_index == len(binary_message):
                new_img.putpixel((x, y), pixel)
                continue

            # Convert the pixel value to binary
            pixel_binary = format(pixel, '08b')

            # Modify the least significant bit of the pixel's intensity value with the message bit
            pixel_binary = pixel_binary[:-1] + binary_message[message_index]
            message_index += 1

            # Convert the modified binary back to an integer
            new_pixel = int(pixel_binary, 2)

            # Set the new pixel in the new image
            new_img.putpixel((x, y), new_pixel)

    # Save the new image with the hidden message
    new_img.save('hidden_grayscale.png')

# Example usage
hide_message("/content/kerberos.jpg", "CSE447")
```

```
def extract_message(image_path):
    # Open the grayscale image
    img = Image.open(image_path).convert('L')

    # Get the size of the image
    width, height = img.size

    binary_message = ''

    # Loop through each pixel in the image
    for x in range(width):
        for y in range(height):

            # Get the pixel value (grayscale intensity)
            pixel = img.getpixel((x, y))

            # Convert the pixel value to binary
            pixel_binary = format(pixel, '08b')

            # Append the least significant bit of the pixel to the binary_message
            binary_message += pixel_binary[-1]

    # Convert the binary message to text
    message = ''
    for i in range(0, len(binary_message), 8):
        byte = binary_message[i:i+8]
        # Check if the byte length is 8 to avoid incomplete bytes
        if len(byte) == 8:
            message += chr(int(byte, 2))

    return message
```
